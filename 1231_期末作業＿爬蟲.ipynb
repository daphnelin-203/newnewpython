### ● 想解決的問題：
#### 近期剛過 1111 購物狂歡節或是 BFCM，看見琳瑯滿目的折扣商品列表，但是預算有限，如何把預算花在刀口上，讓購物「省最大」？

### ● 解決方式：
#### 利用 Python 將折扣頁面爬下來，並經過數據整理與排序，優先顯示折扣優惠最大的商品。



```python
import requests
res = requests.get('https://www.tmall.hk/wh/tmall/import/act/jhb')
```

```python
res
```

### 步驟一：將折扣頁面先爬下來

```python
res.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'lxml')
```

### 步驟二：尋找與金額有關的區塊，擷取 class 名稱，轉入 json 檔

```python
posts = soup.find('textarea', class_='J_dynamic_data')
```

```python
import json
jd = json.loads(posts.text)
```

```python
jd
```

```python
jd.keys()
```

```python
jd['defaultData']
```

### 步驟三：import pandas，讓 pandas 協助將數據整理為表格

```python
import pandas
df = pandas.DataFrame(jd['defaultData'])
```

```python
df
```

### 步驟四：美化 pandas 表格（去除不必要欄位、將欄位名稱修改）

```python
df = df.drop(['brandLogo', 'btnText', 'couponUrl', 'couponValue', 'iconImg', 'itemDescTitle', 'itemImg', 'itemActPrice', ], axis=1)
```

```python
df.rename(columns={'itemActPriceTitle':'價格類型', 'itemDesc1':'商品敘述1', 'itemDesc2':'商品敘述2', 'itemDesc3':'商品敘述3', 'itemDiscount':'折扣價', 'itemMPrice':'折扣後金額', 'itemTagPrice':'原價', 'itemTitle':'產品名稱', 'itemUrl':'產品連結'}, inplace = True)
df
```

### 步驟五：將原價與折扣價相減，得出優惠金額

```python
df['折扣後金額'] = df['折扣後金額'].astype(float)
df['原價'] = df['原價'].astype(float)
df['折扣價'] = df['原價'] - df['折扣後金額']
df
```

### 步驟六：將優惠金額做排序，以方便查看「何者商品折扣最大」
#### 詳見欄位[折扣價] 可得知由大而小的優惠金額

```python
df.sort_values('折扣價', ascending=False)
```
